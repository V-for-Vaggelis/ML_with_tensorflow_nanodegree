# Oxford flowers classifier

In this project, we train an image classifier to recognize different species of flowers. We'll be using [this dataset](http://www.robots.ox.ac.uk/~vgg/data/flowers/102/index.html) from Oxford of 102 flower categories.

The project is broken down into multiple steps:

* Load the image dataset and create a pipeline.
* Build and Train the image classifier on this dataset.
* Use the trained model to perform inference on flower images.

After we've trained our model we'll bind this to a command line application.

## Import Resources

```python
# necessary imports.

import numpy as np
import json
import matplotlib.pyplot as plt
import tensorflow as tf
import tensorflow_datasets as tfds
import tensorflow_hub as hub
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from PIL import Image

import warnings
warnings.filterwarnings('ignore')
```

## Load the Dataset

Let's use `tensorflow_datasets` to load the [Oxford Flowers 102 dataset](https://www.tensorflow.org/datasets/catalog/oxford_flowers102). This dataset has 3 splits: `'train'`, `'test'`, and `'validation'`. 

```python
# Load the dataset with TensorFlow Datasets.
ds, ds_info = tfds.load('oxford_flowers102', as_supervised=True, with_info=True)

training_set = ds['train']
validation_set = ds['validation']
test_set = ds['test']
```

## Explore the Dataset

```python
print(ds_info)
```

```python
# number of examples in each set from the dataset info.
train_num = ds_info.splits['train'].num_examples
validation_num = ds_info.splits['validation'].num_examples
test_num = ds_info.splits['test'].num_examples
# number of classes in the dataset from the dataset info.
print("train number: {}, validation number {}, test number {}".format(train_num, validation_num, test_num))
print("Number of classes: {}".format(ds_info.features['label'].num_classes))
```

We also need to load in a mapping from label to category name. We can find this in the file `label_map.json`.

```python
with open('label_map.json', 'r') as f:
    class_names = json.load(f)
    
print(class_names)
```

## Create Pipeline

```python
# Create a pipeline for each set.
def norm_resize(image, label): 
    image = tf.cast(image, tf.float32)
    image /= 255 # normalize
    image = tf.image.resize(image, [224, 224]) # resize to size used on the pre-trsained networks
    return image, label

batch_size = 32
training_batches = training_set.cache().shuffle(train_num//4).map(norm_resize).batch(batch_size).prefetch(1) # map before batch to avoid errors
validation_batches = validation_set.cache().map(norm_resize).batch(batch_size).prefetch(1)
test_batches = test_set.cache().map(norm_resize).batch(batch_size).prefetch(1)
```

```python
print(training_set)
tf.print(training_batches)
```

# Build and Train the Classifier

Now that the data is ready, it's time to build and train the classifier. We will use the MobileNet pre-trained model from TensorFlow Hub to get the image features. We will build and train a new feed-forward classifier using those features.

In particular we will:

* Load the MobileNet pre-trained network from TensorFlow Hub.
* Define a new, untrained feed-forward network as a classifier.
* Train the classifier.
* Plot the loss and accuracy values achieved during training for the training and validation set.
* Save your trained model as a Keras model. 

```python
# Build and train the network.
mobilenet_v2 = "https://tfhub.dev/google/tf2-preview/mobilenet_v2/feature_vector/4"
mobile_net_layers = hub.KerasLayer(mobilenet_v2, input_shape=(224,224,3))
mobile_net_layers.trainable = False # freeze all mobile net layers (untrainable)

neural_net = tf.keras.Sequential([
  mobile_net_layers,
#   tf.keras.layers.Dropout(0.3),
  tf.keras.layers.Dense(102,activation='softmax') # add a dense layer to with 102 nodes to match the number of classes
])

neural_net.summary() # print our model
```

```python
neural_net.compile(
  optimizer='adam',
  loss='sparse_categorical_crossentropy',
  metrics='accuracy')

EPOCHS = 6
```

```python
model_fit = neural_net.fit(training_batches, epochs=EPOCHS, validation_data=validation_batches)
```

```python
# TODO: Plot the loss and accuracy values achieved during training for the training and validation set.
acc = model_fit.history['accuracy'] # training accuracy
val_acc = model_fit.history['val_accuracy'] # validation accuracy

loss = model_fit.history['loss'] # training loss
val_loss = model_fit.history['val_loss'] # validation loss

epochs_range = range(EPOCHS)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')
plt.ylabel("Accuracy")
plt.xlabel("Epoch")

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.ylabel('Loss')
plt.xlabel("Epoch")
plt.show()
```

## Testing the Network

It's good practice to test the trained network on test data, images the network has never seen either in training or validation. This will give us a good estimate for the model's performance on completely new images.

```python
# Print the loss and accuracy values achieved on the entire test set.

test_score=neural_net.evaluate(test_batches)
print('test loss, test acc:', test_score)
```

## Save the Model

Now that the network is trained, let's save the model so we can load it later for making inference.

```python
# Save the trained model as a Keras model.
path = "./flower_classifier.h5"
neural_net.save(path)
```

## Load the Keras Model

Load the Keras model we saved above.

```python
# Load the Keras model

loaded = tf.keras.models.load_model(path, custom_objects={'KerasLayer': hub.KerasLayer})
loaded.summary()
```

# Inference for Classification

We need to create a function for prediction. But first, we need a function that handles the necessary pre-processing transformations.

```python
# the process_image function, necessary transforms

def process_image(img):
    image = np.squeeze(img)
    image = tf.image.resize(image, (224, 224))/255.0
    return image
```

To check our `process_image` function we have 4 images in the `./test_images/` folder:

* cautleya_spicata.jpg
* hard-leaved_pocket_orchid.jpg
* orange_dahlia.jpg
* wild_pansy.jpg

The code below loads one of the above images using `PIL` and plots the original image alongside the image produced by the `process_image` function. 

```python
image_path = './test_images/hard-leaved_pocket_orchid.jpg'
im = Image.open(image_path)
test_image = np.asarray(im)

processed_test_image = process_image(test_image)

fig, (ax1, ax2) = plt.subplots(figsize=(10,10), ncols=2)
ax1.imshow(test_image)
ax1.set_title('Original Image')
ax2.imshow(processed_test_image)
ax2.set_title('Processed Image')
plt.tight_layout()
plt.show()
```

### Inference

Now let's write a predict function for inference

```python
# the predict function

def predict(im_path, model, top_k):
    im = Image.open(im_path)
    test_image = np.asarray(im) # convert objet to np array
    test_image_proc = process_image(test_image)
    prediction = model.predict(np.expand_dims(test_image_proc, axis=0))
    top_probs, top_indices = tf.math.top_k(prediction, top_k) # keep only k best
    print("Top propabilities",top_probs.numpy()[0])
    top_classes = [class_names[str(value+1)] for value in top_indices.numpy()[0]] # +1 cause labels range from 1 to 102
    print('Top classes', top_classes)
    return top_probs.numpy()[0], top_classes
```

# Sanity Check

It's always good to check the predictions made by your model to make sure they are correct. To check our predictions we will use the 4 images in the `./test_images/` folder:

* cautleya_spicata.jpg
* hard-leaved_pocket_orchid.jpg
* orange_dahlia.jpg
* wild_pansy.jpg

In the cell below we use `matplotlib` to plot the input image alongside the probabilities for the top 5 classes predicted by our model.

```python
# Plot the input image along with the top 5 classes
import glob # to parse directory
infer_images = glob.glob('./test_images/*.jpg')
for im_path in infer_images:
    im = Image.open(im_path)
    test_image = np.asarray(im)
    test_image_proc = process_image(test_image)
    probs, classes = predict(im_path, loaded, 5)
    fig, (ax1, ax2) = plt.subplots(figsize=(12,4), ncols=2)
    ax1.imshow(test_image_proc)
    ax2 = plt.barh(classes[::-1], probs[::-1])
    plt.tight_layout()
    plt.show()
```
